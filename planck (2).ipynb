```python
import matplotlib.pyplot as plt
import numpy as np
import math 

h = 6.626e-34
c = 3.0e+8
k = 1.38e-23

def planck(wav, T):
    a = 2.0*h*c**2
    b = h*c/(wav*k*T)
    intensity = a/ ( (wav**5) * (np.exp(b) - 1.0) )
    return intensity
# generate x-axis in increments from 1nm to 3 micrometer in 1 nm increments
# starting at 1 nm to avoid wav = 0, which would result in division by zero.
#wavelengths = np.arange(1e-9, 3e-6, 1e-3)
wavelengths = np.arange(1e-7,1e-4,1e-9) 
intensity6000 = planck(wavelengths, 6000.)/max(planck(wavelengths, 6000.))
intensity4000 = planck(wavelengths, 4000.)/max(planck(wavelengths, 4000.))
intensity2000 = planck(wavelengths, 2000.)/max(planck(wavelengths, 2000.))
intensity1000 = planck(wavelengths, 1000.)/max(planck(wavelengths, 1000.))
intensity800 = planck(wavelengths, 800.)/max(planck(wavelengths, 800.))
intensity500 = planck(wavelengths, 500.)/max(planck(wavelengths, 500.))
intensity288 = planck(wavelengths, 288.)/max(planck(wavelengths, 288.))
intensity255 = planck(wavelengths, 255.)/max(planck(wavelengths, 255.))


plt.plot(wavelengths/10e-6, intensity6000, 'y-')
plt.plot(wavelengths/10e-6, intensity4000, 'g-') 
plt.plot(wavelengths/10e-6, intensity2000, 'm-') 
plt.plot(wavelengths/10e-6, intensity1000, 'k-') 
plt.plot(wavelengths/10e-6, intensity800, 'c-')
plt.plot(wavelengths/10e-6, intensity500, 'r-')
plt.plot(wavelengths/10e-6, intensity288, 'b-')
plt.plot(wavelengths/10e-6, intensity255, 'b-')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.xscale("log")

plt.grid()
#plt.figure(figsize=(20, 20), dpi=100)
#plt.figsize(3,3)
#plt.axis('scaled')
# show the plot
plt.show()
```
